```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
%matplotlib notebook
```

```python
classes = ["SO4","GOL","EDO","HEM","NAG"]
```

```python
count_train = [59940,80250,46530,50355,14820]
count_train = [int(v/15) for v in count_train]
```

```python
count_test = [443,592,356,369,105]
```

```python
fig, ax = plt.subplots()
ax.bar(classes, count_train, label="Training set")
ax.bar(classes, count_test, label="Test set")
ax.set_ylabel('Number of Examples')
ax.legend()
plt.show()
```

```python
fig.savefig("cnn-class-balance.png")
```
